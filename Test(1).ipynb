# Task 1- Python Programming
# Write Python code to perform the following:
# Problem 1 — Frequency Counter
# Given a list of numbers, return the top 3 most frequent numbers.
# Example
# Input
# [1,2,2,3,3,3,4,4,4,4]
# Output
# 4 --4 times
# 3 -- 3 times
# 2 -- 2 times

```python
nums = [1,2,2,3,3,3,4,4,4,4]
freq = {}

for n in nums:
    if n in freq:
        freq[n] += 1
    else:
        freq[n] = 1
sorted_freq = sorted(freq.items(), key=lambda x: x[1], reverse=True)
for num, count in sorted_freq[:3]:
    print(num, "--", count, "times")
```

# Problem 2 — Dictionary Transformation
# Given the dictionary:
# data = {
# &quot;Aman&quot;: [85, 90, 78],
# &quot;Eakam”: [92, 88, 95],
# &quot;Kiran&quot;: [70, 75, 80]
# }
# Write a program that:
# Calculates the average score of each student
# Returns the student with the highest average
# Expected Output Example
Aman: 84.3
Eakam: 91.6
Kiran: 75
Top Student: Eakam

```python
data = {
    "Aman": [85, 90, 78],
    "Eakam": [92, 88, 95],
    "Kiran": [70, 75, 80]
}
avg_scores = {}
for student, marks in data.items():
    avg = sum(marks) / len(marks)
    avg_scores[student] = avg
    print(f"{student}: {round(avg,1)}")

top_student = max(avg_scores, key=avg_scores.get)
print("Top Student:", top_student)
```
